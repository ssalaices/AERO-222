```python
#3a
def Cube(x):
    return x**3

CubedList = []
for x in range(1, 100001):
    CubedList.append(Cube(x))
```

```python
#3b
def CubeList(numbers):
    CubedList = [num**3 for num in numbers]
    return CubedList      

numbers = [i for i in range(1, 100001)]  

CubedList = CubeList(numbers)
```

**Question 4: Plotting**

Write a code that plots the 3 functions below on the same plot for $x = [0, 1]$

$f_1(x) = \sin(x)$

$f_2(x)  =\cos(x)$

$f_3(x)  = f_1(x)^2 + f_2(x)^2 $

*   Make sure to include a legend on the plot indicating each function.
*   Label the vertical axis *Functions* and the horizontal axis *x*.
*   Use a font size of 12 pt for the legend and 14 pt for the axes.
*   To ensure a smooth plot, include at least a 100 values between $x = 0 $ and $x = 1$.
*   Ensure that each function has a different color.



```python
import numpy as np
from matplotlib import pyplot as plt
from math import sin, cos

def f1(x):
    return sin(x)
    
def f2(x):
    return cos(x)

def f3(x):
    return (f1(x)**2 + f2(x)**2)

xvals = np.linspace(0, 1, 100)

f1vals = [f1(x) for x in xvals]
f2vals = [f2(x) for x in xvals]
f3vals = [f3(x) for x in xvals]

plt.plot(xvals, f1vals)
plt.plot(xvals, f2vals)
plt.plot(xvals, f3vals)

plt.xlabel('x', fontsize = 14)
plt.ylabel('Functions', fontsize = 14)
plt.legend(["f1(x)", "f2(x)", "f3(x)"], fontsize = 12)

plt.show()
```
