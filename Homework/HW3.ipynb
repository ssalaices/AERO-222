#### **Part 1 - Colon notation**

```python
# 1. Vector dot vector: A.B
def Dot(A, B):
    sum = 0
    for i in range(len(A)):
        sum += A[i] * B[i]
    return sum

# 2. Vector x vector: A cross B
def Cross(A, B):
    cross = [
    A[1]*B[2] - B[1]*A[2],
    -(A[0]*B[2] - A[2]*B[0]),
    A[0]*B[1] - A[1]*B[0]
    ]
    return cross

# 3. Matrix*Vector: CA
def MatrixVector(M, V):
    result = np.zeros(V.shape[0])
    for rowNum in range(M.shape[0]):
        for colNum in range(M.shape[0]):
            result[rowNum] = result[rowNum] + M[rowNum, colNum] * V[colNum]
    return result

# 4. Matrix*Matrix: CD
def MatrixMatrix(A, B):
    result = np.zeros((A.shape[0], B.shape[1]))
    for i in range(A.shape[0]):
        for j in range(B.shape[1]):
            for k in range(A.shape[1]):
                result[i, j] += A[i, k] * B[k, j]
    return(result)

print(f"The dot product A.B = {Dot(A, B)}")
print(f"The cross product AxB is: {Cross(A, B)}")
print(f"Matrix * Vector: [C][A] = {MatrixVector(C, A)}")
print(f"Matrix*Matrix: [C][D] = \n{MatrixMatrix(C, D)}")
```
